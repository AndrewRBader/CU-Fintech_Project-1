```python
import pandas as pd
import hvplot.pandas
```

```python
sandp500_df = pd.read_csv("../Resources/S&P500.csv", index_col='Unnamed: 0')

sandp500_df.head()
```

```python
# getting rid of commas

for column in sandp500_df.columns[0::]:
    sandp500_df[column] = sandp500_df[column].replace(",", "", regex=True)
    sandp500_df[column] = sandp500_df[column].replace("-", "", regex=True)
    sandp500_df[column] = pd.to_numeric(sandp500_df[column])


display(sandp500_df.head())
```

```python
sandp500_df.hvplot(
    # setting axes data
    x=0, y='Close',
    # setting axes labels
    xlabel='Date', ylabel='Close',
    # title
    title='S&P 500 Close',
)
```
