```python
import pandas as pd

csv = pd.read_csv("../Resources/S&P500.csv")

csv.head()
```

```python
# getting rid of commas
csv["Open"] = csv["Open"].replace(",", "", regex=True)
csv["High"] = csv["High"].replace(",", "", regex=True)
csv["Low"] = csv["Low"].replace(",", "", regex=True)
csv["Close"] = csv["Close"].replace(",", "", regex=True)
csv["Adj Close"] = csv["Adj Close"].replace(",", "", regex=True)
csv["Volume"] = csv["Volume"].replace(",", "", regex=True)

# getting rid of dashes
csv["Open"] = csv["Open"].replace("-", "", regex=True)
csv["High"] = csv["High"].replace("-", "", regex=True)
csv["Low"] = csv["Low"].replace("-", "", regex=True)
csv["Close"] = csv["Close"].replace("-", "", regex=True)
csv["Volume"] = csv["Volume"].replace("-", "", regex=True)

display(csv.head())

csv['Volume'].dtypes
```

```python
csv["Open"] = pd.to_numeric(csv["Open"])
csv["High"] = pd.to_numeric(csv["High"])
csv["Low"] = pd.to_numeric(csv["Low"])
csv["Close"] = pd.to_numeric(csv["Close"])
csv["Adj Close"] = pd.to_numeric(csv["Adj Close"])
csv["Volume"] = pd.to_numeric(csv["Volume"])

display(csv.head())

csv['Volume'].dtypes
```
